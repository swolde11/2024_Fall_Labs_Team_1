```python
import numpy as np
import matplotlib.pyplot as plt

# 1
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x)

plt.plot(x, y)

# 2
plt.xlabel('x')
plt.ylabel('y')
plt.title('y= Sine(x)')
plt.show()
```

```python
# 3

from mpl_toolkits.mplot3d import axes3d

x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X, Y = np.meshgrid(x, y)
z = np.sqrt(X**2 + Y**2)
Z = np.sin(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_zlim(-5, 5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = sin(sqrt(x^2 + y^2))')

plt.show()
```

```python
import scipy

# 1 
a = np.matrix([[3, 1],[1, 2]])
b = np.matrix([[9], [8]])

x = scipy.linalg.solve(a, b)
print(x)

# 2 
func = lambda u : u**2 + 2*u
res = scipy.optimize.minimize(func, 0)
print(res)

# 3
N = 500
T = 1.0 / 500.0
x = np.linspace(0, N*T, N, endpoint=False)
y = np.sin(100.0*np.pi*x) + 0.5*np.sin(160.0*np.pi*x)

Y = scipy.fft.fft(y)
freqs = scipy.fft.fftfreq(N, T)[:N//2]

plt.plot(freqs, (2.0/N)*np.abs(Y[0:N//2]))
```

```python
import cv2

# 1
image = cv2.imread('goose.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('goose_gray.jpg', gray_image)

# 2
image = cv2.imread('servo.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(gray_image, (3,3), 0) 

edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200)
cv2.imwrite('servo_edge_detect.jpg', edges)
```

```python
# 3
image = cv2.imread('faces.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces_rect = face_cascade.detectMultiScale(gray_image, 1.1, 9) 

for (x, y, w, h) in faces_rect: 
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), thickness=2) 
  
cv2.imwrite('faces_detected.jpg', image)
```
